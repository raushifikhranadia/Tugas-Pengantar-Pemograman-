```python
import numpy as np # pustaka numerik untuk array dan fungsi matematika
from scipy import integrate # pustaka untuk metode integrasi numerik

def integrate_rec(f, a, b, n): # fungsi untuk menghitung integral menggunakan metode rectangle
    # Implementation of the rectanlge method 
    h = (b-a)/n # lebar setiap subinterval
    x = np.linspace(a, b, n+1) # titik-titik pembagi interval [a, b] menjadi n subinterval
    i=0 # indeks untuk iterasi
    area=0 # variabel untuk menyimpan hasil penjumlahan luas persegi panjang
    while i<n: # iterasi untuk menghitung luas persegi panjang pada setiap subinterval
        sup_rect =  f(x[i])*h # luas persegi panjang dengan tinggi f(x[i]) dan lebar h
        area += sup_rect # menambahkan luas persegi panjang ke total area
        i += 1 # increment indeks untuk subinterval berikutnya
    return area # mengembalikan hasil integral yang dihitung menggunakan metode rectangle
'''
We test the Rectangle method on the sine funcion were the definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_5 = integrate_rec(np.sin, 0, np.pi/2, 5) # menghitung integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode rectangle dan n=5
sup_10 = integrate_rec(np.sin, 0, np.pi/2, 10) # menghitung integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode rectangle dan n=10
sup_100 = integrate_rec(np.sin, 0, np.pi/2, 100) # menghitung integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode rectangle dan n=100
sup_12 = integrate_rec (np.sin, 0, np.pi/2, 12) # menghitung integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode rectangle dan n=12

print('Using n=5, the rectangle method returns a value of {:.2f}'.format(sup_5)) # mencetak hasil integral yang dihitung dengan n=5, diformat dengan 2 angka desimal
print('Using n=10, the rectangle method returns a value of {:.2f}'.format(sup_10)) # mencetak hasil integral yang dihitung dengan n=10, diformat dengan 2 angka desimal
print('Using n=100, the rectangle method returns a value of {:.2f}'.format(sup_100)) # mencetak hasil integral yang dihitung dengan n=100, diformat dengan 2 angka desimal
print('Using n=12, the rectangle method returns a value of {:.2f}'.format(sup_12)) # mencetak hasil integral yang dihitung dengan n=12, diformat dengan 2 angka desimal
     
'''
Output:
Using n=5, the rectangle method returns a value of 0.83
Using n=10, the rectangle method returns a value of 0.92
Using n=100, the rectangle method returns a value of 0.99
'''
 
```

```python
import numpy as np # pustaka numerik untuk array dan fungsi matematika

def integrate_trap(f, a, b, n): # fungsi untuk menghitung integral menggunakan metode trapezoidal
    # Implementation of the trapezoidal rule 
    h = (b-a)/n # lebar setiap subinterval
    x = np.linspace(a, b, n+1) # titik-titik pembagi interval [a, b] menjadi n subinterval
    i=1 # indeks untuk iterasi
    area = h*(f(x[0]) + f(x[n]))/2 # luas trapezoid pertama
    while i<n: # iterasi untuk menghitung luas trapezoid pada setiap subinterval
        sup_rect = f(x[i])*h # luas trapezoid dengan tinggi f(x[i]) dan lebar h
        area += sup_rect # menambahkan luas trapezoid ke total area
        i += 1 # increment indeks untuk subinterval berikutnya
    return area # mengembalikan hasil integral yang dihitung menggunakan metode trapezoidal

'''
We test the trapezoidal rule on the known sine funcion were the  
definite integral in the interval [0, π/2] is equal to 1. 
'''

sup_5 = integrate_trap(np.sin, 0, np.pi/2, 5) # menghitung integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode trapezoidal dan n=5
sup_10 = integrate_trap(np.sin, 0, np.pi/2, 10) # menghitung integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode trapezoidal dan n=10
sup_56 = integrate_trap(np.sin, 0, np.pi/2, 56) # menghitung integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode trapezoidal dan n=56

print('Using n=5, the trapezoidal rule returns a value of {:.2f}'.format(sup_5)) # mencetak hasil integral yang dihitung dengan n=5, diformat dengan 2 angka desimal
print('Using n=10, the trapezoidal rule returns a value of {:.2f}'.format(sup_10)) # mencetak hasil integral yang dihitung dengan n=10, diformat dengan 2 angka desimal
print('Using n=56, the trapezoidal rule returns a value of {:.2f}'.format(sup_56)) # mencetak hasil integral yang dihitung dengan n=56, diformat dengan 2 angka desimal
   
'''
Output:
Using n=5, the trapezoidal rule returns a value of 0.99
Using n=10, the trapezoidal rule returns a value of 1.00
Using n=56, the trapezoidal rule returns a value of 1.00
'''
 
```

```python
import numpy as np # pustaka numerik untuk array dan fungsi matematika
import matplotlib.pyplot as plt # pustaka untuk membuat grafik dan visualisasi data

def density(): # fungsi untuk menghitung densitas pada setiap lapisan interior bumi, dengan menggunakan data densitas yang diketahui untuk setiap lapisan
    ro_inner_core = np.linspace(13100, 12800, 1220) # array yang berisi densitas pada lapisan inti dalam bumi, mulai dari 13100 kg/m^3 pada jarak 1 km hingga 12800 kg/m^3 pada jarak 1220 km dengan interval yang sesuai
    ro_outer_core = np.linspace(12200, 9900, 2259) # array yang berisi densitas pada lapisan inti luar bumi, mulai dari 12200 kg/m^3 pada jarak 1220 km hingga 9900 kg/m^3 pada jarak 3480 km dengan interval yang sesuai
    ro_lower_mantle = np.linspace(5600,4400,2171) # array yang berisi densitas pada lapisan mantel bawah bumi, mulai dari 5600 kg/m^3 pada jarak 3480 km hingga 4400 kg/m^3 pada jarak 5651 km dengan interval yang sesuai
    ro_upper_mantle = np.linspace(4130,3400,720) # array yang berisi densitas pada lapisan mantel atas bumi, mulai dari 4130 kg/m^3 pada jarak 5651 km hingga 3400 kg/m^3 pada jarak 6371 km dengan interval yang sesuai
    ro_crust = np.linspace(3100,2700,30) # array yang berisi densitas pada lapisan kerak bumi, mulai dari 3100 kg/m^3 pada jarak 6371 km hingga 2700 kg/m^3 pada jarak 6400 km dengan interval yang sesuai
    return np.concatenate((ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust)) # menggabungkan semua array densitas dari setiap lapisan menjadi satu array final yang berisi densitas pada setiap jarak dari pusat bumi hingga permukaan bumi, kemudian mengembalikan array tersebut

def trapezoid_manual(y, x): # fungsi untuk menghitung integral menggunakan metode trapezoidal secara manual, dengan input y sebagai nilai fungsi pada titik-titik pembagi x
    h = np.diff(x) # menghitung lebar setiap subinterval dengan menggunakan fungsi np.diff untuk menghitung selisih antara elemen-elemen dalam array x
    return np.sum((y[:-1] + y[1:])/2 * h) # menghitung luas trapezoid pada setiap subinterval dengan menggunakan rumus (y[i] + y[i+1])/2 * h[i], kemudian menjumlahkan semua luas trapezoid untuk mendapatkan hasil integral

r = np.linspace(1,6400,6400) # array yang berisi jarak dari pusat bumi ke permukaan bumi, mulai dari 1 km hingga 6400 km dengan interval 1 km
ro = density() # array yang berisi densitas pada setiap jarak dari pusat bumi hingga permukaan bumi, dengan menggunakan fungsi density yang telah didefinisikan sebelumnya

def gravity(r, ro): # fungsi untuk menghitung percepatan gravitasi pada setiap jarak dari pusat bumi hingga permukaan bumi, dengan menggunakan rumus g = (4 * pi * G / r^2) * integral(ro(r') * r'^2 dr' dari 0 hingga r), dimana G adalah konstanta gravitasi, r adalah jarak dari pusat bumi, dan ro(r') adalah densitas pada jarak r'
    g = np.zeros(len(r)) # array untuk menyimpan nilai percepatan gravitasi pada setiap jarak dari pusat bumi hingga permukaan bumi, diinisialisasi dengan nol
    Gr = 6.67408e-11 # konstanta gravitasi dalam satuan m^3 kg^-1 s^-2
    r_m = r * 1000 # mengkonversi jarak dari pusat bumi ke permukaan bumi dari satuan km ke satuan meter dengan mengalikan setiap elemen dalam array r dengan 1000
    
    for i in range(1,len(r)): # iterasi untuk menghitung percepatan gravitasi pada setiap jarak dari pusat bumi hingga permukaan bumi, dimulai dari indeks 1 karena pada indeks 0 (jarak 1 km) percepatan gravitasi belum dapat dihitung
        r1 = r_m[0:i] # array yang berisi jarak dari pusat bumi hingga jarak r[i], dalam satuan meter
        ro1 = ro[0:i] # array yang berisi densitas pada setiap jarak dari pusat bumi hingga jarak r[i], sesuai dengan array r1
        r2 = r1[-1] # jarak r[i] dalam satuan meter, yaitu elemen terakhir dari array r1
        y = ro1 * r1**2 # array yang berisi nilai fungsi ro(r') * r'^2 pada setiap jarak dari pusat bumi hingga jarak r[i], dengan menggunakan array ro1 dan r1
        y_int = trapezoid_manual(y, r1) # menghitung integral dari fungsi ro(r') * r'^2 dr' dari 0 hingga r[i] dengan menggunakan fungsi trapezoid_manual yang telah didefinisikan sebelumnya, dengan input y sebagai nilai fungsi pada titik-titik pembagi r1
        g[i] = (4 * np.pi * Gr / r2**2) * y_int # menghitung percepatan gravitasi pada jarak r[i] dengan menggunakan rumus g = (4 * pi * G / r^2) * integral(ro(r') * r'^2 dr' dari 0 hingga r), dimana G adalah konstanta gravitasi, r adalah jarak dari pusat bumi, dan ro(r') adalah densitas pada jarak r', kemudian menyimpan hasilnya dalam array g pada indeks i
        
    return g # mengembalikan array g yang berisi nilai percepatan gravitasi pada setiap jarak dari pusat bumi hingga permukaan bumi

g = gravity(r, ro) # menghitung percepatan gravitasi pada setiap jarak dari pusat bumi hingga permukaan bumi dengan menggunakan fungsi gravity yang telah didefinisikan sebelumnya, dengan input r sebagai array jarak dan ro sebagai array densitas

fig, ax = plt.subplots() # membuat objek figure dan axes untuk membuat grafik
ax.plot(r, g, 'b-') # membuat grafik dengan sumbu x adalah jarak dari pusat bumi (r) dan sumbu y adalah percepatan gravitasi (g), dengan menggunakan garis berwarna biru
ax.set_xlabel('Distance from the Earth center r [km]') # memberikan label pada sumbu x dengan satuan km, menggunakan r sebagai simbol untuk jarak dari pusat bumi
ax.set_ylabel(r'g [m/s$^2$]') # memberikan label pada sumbu y dengan satuan m/s^2, menggunakan g sebagai simbol untuk percepatan gravitasi
ax.grid(True) # menambahkan grid pada grafik untuk memudahkan pembacaan nilai pada grafik
ax.set_title("Gravitational Acceleration inside Earth") # memberikan judul pada grafik untuk menjelaskan apa yang ditampilkan pada grafik
plt.show() # menampilkan grafik yang telah dibuat, yang menunjukkan percepatan gravitasi pada setiap jarak dari pusat bumi hingga permukaan bumi
```

```python
import numpy as np# pustaka numerik untuk array dan fungsi matematika
import matplotlib.pyplot as plt # pustaka untuk membuat grafik dan visualisasi data

def trapezoid_manual(y, x): # fungsi untuk menghitung integral menggunakan metode trapezoidal secara manual, dengan input y sebagai nilai fungsi pada titik-titik pembagi x
    h = np.diff(x) # menghitung lebar setiap subinterval dengan menggunakan fungsi np.diff untuk menghitung selisih antara elemen-elemen dalam array x
    return np.sum((y[:-1] + y[1:])/2 * h) # menghitung luas trapezoid pada setiap subinterval dengan menggunakan rumus (y[i] + y[i+1])/2 * h[i], kemudian menjumlahkan semua luas trapezoid untuk mendapatkan hasil integral

def pressure(r, ro, g): # fungsi untuk menghitung tekanan litostatik pada setiap jarak dari pusat bumi hingga permukaan bumi, dengan menggunakan rumus P(r) = integral(ro(r') * g(r') dr' dari r hingga permukaan bumi), dimana ro(r') adalah densitas pada jarak r', g(r') adalah percepatan gravitasi pada jarak r', dan r adalah jarak dari pusat bumi
    p = np.zeros(len(r)) # array untuk menyimpan nilai tekanan litostatik pada setiap jarak dari pusat bumi hingga permukaan bumi, diinisialisasi dengan nol
    r_m = r * 1000 # km -> m
    
    for i in range(len(r)): # iterasi untuk menghitung tekanan litostatik pada setiap jarak dari pusat bumi hingga permukaan bumi 
        r1 = r_m[i:] # array yang berisi jarak dari pusat bumi hingga permukaan bumi, dimulai dari jarak r[i] hingga jarak 6400 km, dalam satuan meter
        ro1 = ro[i:] # array yang berisi densitas pada setiap jarak dari pusat bumi hingga permukaan bumi, dimulai dari jarak r[i] hingga jarak 6400 km, sesuai dengan array r1  
        g1 = g[i:] # array yang berisi percepatan gravitasi pada setiap jarak dari pusat bumi hingga permukaan bumi, dimulai dari jarak r[i] hingga jarak 6400 km, sesuai dengan array r1    
        y = ro1 * g1 # array yang berisi nilai fungsi ro(r') * g(r') pada setiap jarak dari pusat bumi hingga permukaan bumi, dengan menggunakan array ro1 dan g1    
        p[i] = trapezoid_manual(y, r1) # menghitung integral dari fungsi ro(r') * g(r') dr' dari r[i] hingga permukaan bumi dengan menggunakan fungsi trapezoid_manual yang telah didefinisikan sebelumnya, dengan input y sebagai nilai fungsi pada titik-titik pembagi r1, kemudian menyimpan hasilnya dalam array p pada indeks i
    return p # mengembalikan array p yang berisi nilai tekanan litostatik pada setiap jarak dari pusat bumi hingga permukaan bumi

r = np.linspace(1,6400,6400) # array yang berisi jarak dari pusat bumi ke permukaan bumi, mulai dari 1 km hingga 6400 km dengan interval 1 km
ro = np.linspace(13100, 2700, 6400) # array yang berisi densitas pada setiap jarak dari pusat bumi hingga permukaan bumi, mulai dari 13100 kg/m^3 pada jarak 1 km hingga 2700 kg/m^3 pada jarak 6400 km dengan interval yang sesuai
g = np.linspace(10, 0, 6400) # array yang berisi percepatan gravitasi pada setiap jarak dari pusat bumi hingga permukaan bumi, mulai dari 10 m/s^2 pada jarak 1 km hingga 0 m/s^2 pada jarak 6400 km dengan interval yang sesuai

p = pressure(r, ro, g)/1e9 # menghitung tekanan litostatik pada setiap jarak dari pusat bumi hingga permukaan bumi dengan menggunakan fungsi pressure yang telah didefinisikan sebelumnya, dengan input r sebagai array jarak, ro sebagai array densitas, dan g sebagai array percepatan gravitasi, kemudian membagi hasilnya dengan 1e9 untuk mengkonversi ke satuan GPa
z = np.linspace(6400, 1, 6400) # array yang berisi kedalaman dari permukaan bumi ke pusat bumi, mulai dari 6400 km hingga 1 km dengan interval 1 km, digunakan untuk sumbu x pada grafik tekanan litostatik vs kedalaman

fig, ax = plt.subplots() # membuat objek figure dan axes untuk membuat grafik
ax.plot(z, p, 'r-') # membuat grafik dengan sumbu x adalah kedalaman dari permukaan bumi ke pusat bumi (z) dan sumbu y adalah tekanan litostatik (p), dengan menggunakan garis berwarna merah
ax.set_xlabel('Depth z from the Earth Surface [km]') # memberikan label pada sumbu x dengan satuan km, menggunakan z sebagai simbol untuk kedalaman dari permukaan bumi ke pusat bumi
ax.set_ylabel('P [GPa]') # memberikan label pada sumbu y dengan satuan GPa, menggunakan P sebagai simbol untuk tekanan litostatik
ax.set_title('Lithostatic Pressure vs Depth') # memberikan judul pada grafik untuk menjelaskan apa yang ditampilkan pada grafik, yaitu hubungan antara tekanan litostatik dan kedalaman dari permukaan bumi ke pusat bumi
ax.grid(True) # menambahkan grid pada grafik untuk memudahkan pembacaan nilai pada grafik
plt.show() # menampilkan grafik yang telah dibuat, yang menunjukkan hubungan antara tekanan litostatik dan kedalaman dari permukaan bumi ke pusat bumi
```
