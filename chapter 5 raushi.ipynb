```python
import pandas as pd # untuk membaca data dari file excel
from scipy.stats.mstats import gmean, hmean # untuk menghitung geometric mean dan harmonic mean
import matplotlib.pyplot as plt # untuk membuat histogram dan menampilkan garis vertikal untuk mean 

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca data dari file excel dan menyimpannya dalam variabel my_dataset
a_mean = my_dataset.U.mean()# menghitung arithmetic mean dari kolom U dalam dataset
g_mean = gmean(my_dataset['U'])# menghitung geometric mean dari kolom U dalam dataset
h_mean = hmean(my_dataset['U'])# menghitung harmonic mean dari kolom U dalam dataset 

print ('-------')# mencetak garis pemisah
print ('arithmetic mean')# mencetak judul untuk arithmetic mean
print ("{0:.1f} [ppm]".format(a_mean))# mencetak nilai arithmetic mean dengan format satu angka di belakang koma dan satuan ppm
print ('-------')# mencetak garis pemisah

print ('geometric mean')# mencetak judul untuk geometric mean
print ("{0:.1f} [ppm]".format(g_mean))# mencetak nilai geometric mean dengan format satu angka di belakang koma dan satuan ppm
print ('-------')# mencetak garis pemisah

print ('harmonic mean')# mencetak judul untuk harmonic mean
print ("{0:.1f} [ppm]".format(h_mean))# mencetak nilai harmonic mean dengan format satu angka di belakang koma dan satuan ppm
print ('-------')# mencetak garis pemisah

fig, ax = plt.subplots()# membuat figure dan exes untuk plotting
ax.hist(my_dataset.U, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8)# membuat histogram dari kolom U dalam dataset dengan jumlah bin otomatis, menormalkan histogram, memberikan warna tepi pada setiap bin, memberikan label untuk legend, dan mengatur transparansi histogram 
ax.axvline(a_mean, color='red', label='Arithmetic mean', linewidth=2)# menambahkan garis vertikal pada posisi arithmetic mean dengan warna merah, label untuk legenda, dan ketebalan garis 2
ax.axvline(g_mean, color='yellow',  label='Geometric mean', linewidth=2)# menambahkan garis vertikal pada posisi geometric mean dengan warna kuning, label untuk legenda, dan ketebalan garis 2
ax.axvline(h_mean, color='blue',  label='Harmonic mean', linewidth=2)# menambahkan garis vertikal pada posisi harmonic mean dengan warna biru, label untuk legenda, dan ketebalan garis 2
ax.set_xlabel('U [ppm]')# menambahkan label pada sumbu x
ax.set_ylabel('Probability density')# menambahkan label pada sumbu y
ax.legend()# menampilkan legenda pada plot

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd# untuk membaca data dari file excel
import numpy as np# untuk menghitung histogram dan bin edges
import matplotlib.pyplot as plt# untuk membuat histogram dan menampilkan garis vertikal untuk modal value

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')# membaca data dari file excel dan menyimpannya dalam variabel my_dataset

hist, bin_edges = np.histogram(my_dataset['Hf'], bins= 35, density=True)# menghitung histogram dan bin edges dari kolom Hf dalam dataset dengan 35 bin dan menormalkan histogram
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2# menghitung modal value dengan mengambil rata-rata dari bin edges yang memiliki frekuensi tertinggi dalam histogram

print ('modal value: {0:.0f} [ppm]'.format(modal_value))# mencetak nilai modal value dengan format tanpa angka di belakang koma dan satuan ppm

fig, ax = plt.subplots()# membuat figure dan axes untuk plotting
ax.hist(my_dataset.Hf, bins=35, density=True, edgecolor='k', label="Measurements Hist", alpha=0.8) # membuat histogram dari kolom Hf dalam dataset dengan 35 bin, menormalkan histogram, memberikan warna tepi pada setiap bin, memberikan label untuk legend, dan mengatur transparansi histogram
ax.axvline(modal_value, color="blue", label="Modal value", linewidth=2)# menambahkan garis vertikal pada posisi modal value dengan warna biru, label untuk legenda, dan ketebalan garis 2
ax.set_xlabel('Hf [ppm]')# menambahkan label pada sumbu x
ax.set_ylabel('Probability density')# menambahkan label pada sumbu y
ax.legend()# menampilkan legenda pada plot

'''
Output: modal value: 277 [ppm]
'''
```

```python
import pandas as pd# untuk membaca data dari file excel
import matplotlib.pyplot as plt# untuk membuat histogram dan menampilkan garis vertikal untuk standar deviasi

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')# membaca data dari file excel dan menyimpannya dalam variabel my_dataset

variance =  my_dataset['Pb'].var()# menghitung variance dari kolom Pb dalam dataset
stddev =  my_dataset['Pb'].std()# menghitung standar deviasi dari kolom Pb dalam dataset

print ('-------')# mencetak garis pemisah
print ('Variance')# mencetak judul untuk variance
print("{0:.0f} [square ppm]".format(variance))# mencetak nilai variance dengan format tanpa angka di belakang koma dan satuan square ppm
print ('-------')# mencetak garis pemisah
print ('Standard Deviation')# mencetak judul untuk standar deviasi
print("{0:.0f} [ppm]".format(stddev))# mencetak nilai standar deviasi dengan format tanpa angka di belakang koma dan satuan ppm
print ('-------')# mencetak garis pemisah

fig, ax = plt.subplots()# membuat figure dan axes untuk plotting
ax.hist(my_dataset.Pb, bins= 70, density = True, edgecolor='k', label='Measurements Hist')# membuat histogram dari kolom Pb dalam dataset dengan 70 bin, menormalkan histogram, memberikan warna tepi pada setiap bin, memberikan label untuk legend
ax.axvline(my_dataset['Pb'].mean()-stddev, color='pink', label=r'mean - 1$\sigma$', linewidth=2)# menambahkan garis vertikal pada posisi mean - 1 standar deviasi dengan warna pink, label untuk legenda, dan ketebalan garis 2
ax.axvline(my_dataset['Pb'].mean()+stddev, color='purple', label=r'mean + 1$\sigma$', linewidth=2)# menambahkan garis vertikal pada posisi mean + 1 standar deviasi dengan warna ungu, label untuk legenda, dan ketebalan garis 2
ax.axvspan(my_dataset['Pb'].mean()-stddev, my_dataset['Pb'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$')# menambahkan area berwarna oranye antara mean - 1 standar deviasi dan mean + 1 standar deviasi dengan transparansi 0.1 dan label untuk legenda
ax.set_xlabel('Pb [ppm]')# menambahkan label pada sumbu x
ax.set_ylabel('Probability density')# menambahkan label pada sumbu y
ax.legend()# menampilkan legenda pada plot

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
import pandas as pd# untuk membaca data dari file excel
import numpy as np# untuk menghitung persentil
import matplotlib.pyplot as plt# untuk membuat histogram dan menampilkan garis vertikal untuk interquartile range (IQR)

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')# membaca data dari file excel dan menyimpannya dalam variabel my_dataset

q1 = np.percentile(my_dataset.Gd, 25, method= 'midpoint')# menghitung kuartil pertama (Q1) dari kolom Gd dalam dataset dengan metode midpoint 
q3 = np.percentile(my_dataset.Gd, 75, method= 'midpoint')# menghitung kuartil ketiga (Q3) dari kolom Gd dalam dataset dengan metode midpoint
  
iqr = q3 - q1 # Interquaritle range (IQR)  

print ('-------')# mencetak garis pemisah
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr))# mencetak nilai interquartile range (IQR) dengan format tanpa angka di belakang koma dan satuan ppm
print ('-------')# mencetak garis pemisah

fig, ax = plt.subplots()# membuat figure dan axes untuk plotting
ax.hist(my_dataset.Gd, bins='auto', density=True, edgecolor='k', label='Measurements Hist')# membuat histogram dari kolom Gd dalam dataset dengan bin otomatis, menormalkan histogram, memberikan warna tepi pada setiap bin, memberikan label untuk legend
ax.axvline(q1, color='purple', label='Q1', linewidth=2)# menambahkan garis vertikal pada posisi Q1 dengan warna ungu, label untuk legenda, dan ketebalan garis 2
ax.axvline(q3, color='pink', label='Q3', linewidth=2)# menambahkan garis vertikal pada posisi Q3 dengan warna pink, label untuk legenda, dan ketebalan garis 2
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)')# menambahkan area berwarna oranye antara Q1 dan Q3 dengan transparansi 0.1 dan label untuk legenda
ax.set_xlabel('Gd [ppm]')# menambahkan label pada sumbu x
ax.set_ylabel('Probability density')# menambahkan label pada sumbu y
ax.legend()# menampilkan legenda pada plot

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import pandas as pd# untuk membaca data dari file excel
import numpy as np# untuk menghitung histogram dan bin edges
import matplotlib.pyplot as plt# untuk membuat histogram dan menampilkan garis vertikal untuk modal value

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')# membaca data dari file excel dan menyimpannya dalam variabel my_dataset

a_mean = my_dataset.Ta.mean()# menghitung rata-rata dari kolom Ta dalam dataset

median = my_dataset.Ta.median()# menghitung median dari kolom Ta dalam dataset

hist, bin_edges = np.histogram(my_dataset['Ta'], bins=20, density=True)# menghitung histogram dan bin edges dari kolom Ta dalam dataset dengan 20 bin dan menormalkan histogram
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2# menghitung modal value dengan mengambil rata-rata dari bin edges yang memiliki frekuensi tertinggi dalam histogram

fig, ax = plt.subplots()# membuat figure dan axes untuk plotting
ax.hist(my_dataset.Ta, bins=95, density=True, edgecolor='k', label="Measurements Hist")# membuat histogram dari kolom Ta dalam dataset dengan 95 bin, menormalkan histogram, memberikan warna tepi pada setiap bin, memberikan label untuk legend, dan mengatur transparansi histogram 
ax.axvline(modal_value, color='gray', label='Modal Value', linewidth=2)# menambahkan garis vertikal pada posisi modal value dengan warna abu-abu, label untuk legenda, dan ketebalan garis 2
ax.axvline(median, color='purple',  label='Median Value', linewidth=2)# menambahkan garis vertikal pada posisi median dengan warna ungu, label untuk legenda, dan ketebalan garis 2
ax.axvline(a_mean, color='pink',  label='Arithmetic mean', linewidth=2)# menambahkan garis vertikal pada posisi arithmetic mean dengan warna pink, label untuk legenda, dan ketebalan garis 2
ax.set_xlabel('Ta [ppm]')# menambahkan label pada sumbu x
ax.set_ylabel('Probability density')# menambahkan label pada sumbu y
ax.legend()# menampilkan legenda pada plot

```

```python
import numpy as np# untuk menghitung histogram dan bin edges
from scipy.stats import skew# untuk menghitung skewness

a_mean = my_dataset.Tm.mean()# menghitung rata-rata dari kolom Tm dalam dataset
median = my_dataset.Tm.median()# menghitung median dari kolom Tm dalam dataset
hist, bin_edges = np.histogram(my_dataset['Tm'], bins=20, density=True)# menghitung histogram dan bin edges dari kolom Tm dalam dataset dengan 20 bin dan menormalkan histogram
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2# menghitung modal value dengan mengambil rata-rata dari bin edges yang memiliki frekuensi tertinggi dalam histogram
standard_deviation = my_dataset['Tm'].std()# menghitung deviasi standar dari kolom Tm dalam dataset

a1 = (a_mean - modal_value) / standard_deviation# menghitung Pearson's first coefficient of skewness dengan rumus (mean - modal value) / standar deviasi 
a2 = 3 * (a_mean - median) / standard_deviation# menghitung Pearson's 2nd moment of skewness dengan rumus 3 * (mean - median) / standar deviasi
g1 = skew(my_dataset['Tm'])# menghitung Fisher-Pearson's coefficient of skewness menggunakan fungsi skew dari scipy.stats

print ('-------')# mencetak garis pemisah
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1))# mencetak nilai Pearson's first coefficient of skewness dengan format dua angka di belakang koma
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2))# mencetak nilai Pearson's 2nd moment of skewness dengan format dua angka di belakang koma
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))# mencetak nilai Fisher-Pearson's coefficient of skewness dengan format dua angka di belakang koma
print ('-------')# mencetak garis pemisah

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```

```python
import pandas as pd# untuk membaca data dari file excel
import matplotlib.pyplot as plt# untuk membuat histogram dan menampilkan garis vertikal untuk boxplot

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')# membaca data dari file excel dan menyimpannya dalam variabel my_dataset

fig, ax = plt.subplots()# membuat figure dan axes untuk plotting
my_flierprops = dict(markerfacecolor='#f8e9a1', markeredgecolor='#24305e', marker='o')# mengatur properti untuk outliers pada boxplot dengan warna marker kuning muda, warna tepi marker biru tua, dan bentuk marker bulat
my_medianprops = dict(color='#f76c6c', linewidth = 2)# mengatur properti untuk median pada boxplot dengan warna merah muda dan ketebalan garis 2
my_boxprops = dict(facecolor='#a8d0e6', edgecolor='#24305e')# mengatur properti untuk box pada boxplot dengan warna isi biru muda dan warna tepi biru tua
ax.boxplot(my_dataset.La, patch_artist = True, notch=True, flierprops=my_flierprops, medianprops=my_medianprops, boxprops=my_boxprops)# membuat boxplot untuk kolom La dalam dataset dengan meng 
ax.set_ylabel('La [ppm]')# menambahkan label pada sumbu y
ax.set_xticks([1])# mengatur posisi x-ticks pada angka 1
ax.set_xticklabels(['all Epochs'])# mengatur label untuk x-ticks dengan teks 'all Epochs'
plt.show()# menampilkan plot
```
